### Tfrecord data example
For the convenience, we rename the folder of model with Wavelet preprocessing as WT5-train


```python
import tensorflow as tf
import json
from model import Model
from TFrecords_Preparation import encoder_main, Batch_Generation
%pylab inline
from matplotlib.patches import Rectangle
```

```python
filepure = './data/test.tfrecords'
filepure_queue = tf.train.string_input_producer([filepure], num_epochs=None)
reader_pure = tf.TFRecordReader()
_, serialized_pure = reader_pure.read(filepure_queue)

filewave = './data/WT5-test.tfrecords'
filewave_queue = tf.train.string_input_producer([filewave], num_epochs=None)
reader_wave = tf.TFRecordReader()
_, serialized_wave = reader_wave.read(filewave_queue)
```

```python
features = tf.parse_single_example(
  serialized_pure,
  features={
      'image': tf.FixedLenFeature([], tf.string),
      'length': tf.FixedLenFeature([], tf.int64),
      'digits': tf.FixedLenFeature([5], tf.int64)
  })

image = tf.decode_raw(features['image'], tf.uint8)
image = tf.reshape(image, [64, 64, 3])
length = features['length']
digits = features['digits']
sess = tf.InteractiveSession()
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(coord=coord)
```

```python
(image_test, length_test, digits_test) = sess.run([image, length, digits])
print ('length: %d, digits: %d,%d,%d,%d,%d' % (length_test, digits_test[0], digits_test[1], digits_test[2], digits_test[3], digits_test[4]))

plt.imshow(image_test)
plt.show()
coord.request_stop()
coord.join(threads)
sess.close()
```

```python
features = tf.parse_single_example(
  serialized_wave,
  features={
      'image': tf.FixedLenFeature([], tf.string),
      'length': tf.FixedLenFeature([], tf.int64),
      'digits': tf.FixedLenFeature([5], tf.int64)
  })

image = tf.decode_raw(features['image'], tf.uint8)
image = tf.reshape(image, [64, 64, 3])
length = features['length']
digits = features['digits']
sess = tf.InteractiveSession()
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(coord=coord)
```

```python
(image_test, length_test, digits_test) = sess.run([image, length, digits])
print ('length: %d, digits: %d,%d,%d,%d,%d' % (length_test, digits_test[0], digits_test[1], digits_test[2], digits_test[3], digits_test[4]))

fig = plt.figure(figsize=(12, 3))
titles = ['Gray Scale', 'Spectrum', 'Wavelet Transform']
for i in range(3):
    ax = fig.add_subplot(1, 3, i + 1)
    ax.imshow(image_test[:,:,i],cmap=plt.cm.gray)
    ax.set_title(titles[i], fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])

fig.tight_layout()
plt.show()
```

```python
json_file = './data/meta.json'
with open(json_file, 'r') as f:
    content = json.load(f)
    num_train_examples = content['num_examples']['train']
    num_val_examples = content['num_examples']['val']
    num_test_examples = content['num_examples']['test']
print ('Number of Train examples: %d' % num_train_examples)
print ('Number of Validation examples: %d' % num_val_examples)
print ('Number of Test examples: %d' % num_test_examples)
```

### Inference Example

```python
path_to_tfrecords_file = './data/WT5-test.tfrecords'
batch_size = 16
model_param = {
        'filters_num': [48,64,128,160,192,192,192,192],
        'kernel_size': [5,5,5,5,5,5,5,5],
        'pool_size': [2,2,2,2,2,2,2,2],
        'strides': [2,1,2,1,2,1,2,1],
        'dense': [3072,3072],
        'flatten': [4,4]
}
```

```python
image_batch, length_batch, digits_batch = Batch_Generation(path_to_tfrecords_file,num_examples=num_test_examples,batch_size=batch_size,shuffled=False)    
length_logits, digits_logits = Model.inference(image_batch, model_param, drop_rate=0.0)
length_predictions = tf.argmax(length_logits, axis=1)
digits_predictions = tf.argmax(digits_logits, axis=2)
digits_batch_string = tf.reduce_join(tf.as_string(digits_batch), axis=1)
digits_predictions_string = tf.reduce_join(tf.as_string(digits_predictions), axis=1)

sess = tf.InteractiveSession()
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(coord=coord)
```

```python
restorer = tf.train.Saver()
checkpoint_path = tf.train.latest_checkpoint('logs/WT5-train')
restorer.restore(sess, checkpoint_path)

length_predictions_val, digits_predictions_string_val, image_batch_val = sess.run([length_predictions, digits_predictions_string, image_batch])
image_batch_val = (image_batch_val / 2.0) + 0.5
```

```python
fig, axes = plt.subplots(4, 4, figsize=(30, 30))
for i, ax in enumerate(axes.flat): 
    title = 'length: %d\ndigits: %s' % (length_predictions_val[i],digits_predictions_string_val[i])
    ax.imshow(image_batch_val[i][:,:,0],cmap=plt.cm.gray)
    ax.set_title(title)
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
coord.request_stop()
coord.join(threads)
sess.close()
```
